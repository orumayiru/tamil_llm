```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# create a webdriver instance
driver = webdriver.Chrome()

# navigate to the starting url
driver.get(url)

# wait for the page to load
time.sleep(1)

# initialize an empty set to store all links
all_links = set()
ls = []
# function to recursively find links in div with id=texttospeak
def find_links(url):
    driver.get(url)
    # find all links in div with id=texttospeak
    links = driver.find_elements(By.CSS_SELECTOR, "div#texttospeak a")
  
    if links != []:
        # iterate over the links
        for link in links:
            with open("agri_links.txt", "a") as f:
                f.write(link.get_attribute("href") + "\n")
            #print(link.get_attribute("href"))
            #ls.append(link.get_attribute("href"))
            links = find_links(link.get_attribute("href"))
        
find_links(url)


# close the webdriver
driver.quit()
```
